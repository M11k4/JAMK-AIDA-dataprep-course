## Vaihe 1 - some info on recent trains into a DataFrame

#### Tehtävä 1. Hae Digitrafficin REST-rajapintapalvelusta (JSON-formaatissa) aktiiviset junat pääteosoitteella `/live-trains` Helsingin asemalta, joiden hakuparametrit ovat seuraavat:
* arrived_trains=10
* departed_trains=10
* include_nonstopping=False

```python
import requests

junaturl =  "https://rata.digitraffic.fi/api/v1/live-trains"
asema = "/station/HKI"
parametrit = "?arrived_trains=10&departed_trains=10&include_nonstopping=false"

req = requests.get(url=junaturl+asema+parametrit)

#req.status_code

junadata_json = req.json()

print(len(junadata_json))
```

## Vaihe 2 - reading data from a file into the DataFrame using regex

#### Tehtävä 7. Pura tämän kansion sisällä olevan `data.html` -tiedoston `table`-elementin solujen arvot käyttäen regex-syntaksia ja yhdistä ne vaiheessa 1 luotuun dataframeen. Huom: tehtävässä tulee käyttää vain regexiä (Esim. Pandasin read_html()-metodin käyttö on kielletty)!

```python
import re

f = open('data.html', 'r')
txt = f.read()
f.close()

headpattern = r'<th>([^<]*)<\/th>'
headers = re.findall(headpattern,txt)
#headers

rowpattern = r'<tr(?:[^>])*>\s*((?:<td>[^<]*<\/td>\s*)+)'
rows = re.findall(rowpattern,txt)
#rows

#cellpattern = r'<td>([^<]*)<\/td>'
cellpattern = r'<td>\s*([^<\n]*)\s*<\/td>'
#cell = re.findall(cellpattern,rows[0])
#cell

rowed = list(map(lambda row: re.findall(cellpattern,row),rows))
rowed

funadata = pd.DataFrame( rowed, columns=headers )
funadata

punadata = pd.concat([punadata,funadata])
punadata.reset_index(inplace=True,drop=True)
punadata.tail()
```

## Vaihe 3 - converting and saving the DataFrame as an SQLite file

#### Tehtävä 8. Tee SQLite tietokanta ja lisää sinne vaiheen 2 tuloksena saatu dataframe.


```python
import sqlite3

database_connection  = sqlite3.connect("junat.db")
c = database_connection.cursor()

c.execute("""CREATE TABLE if not exists junadata (stationShortCode, stationUICCode, countryCode, type, trainStopping, commercialTrack, cancelled, scheduledTime, actualTime, trainNumber)""")

# stationShortCode, stationUICCode, countryCode, type, trainStopping, commercialTrack, cancelled, scheduledTime, actualTime, TrainNumber

data_rows = list(punadata.itertuples(index=False,name=None))

for data_row in data_rows:
    c.execute('''INSERT INTO junadata VALUES {0}'''.format(data_row))

database_connection.commit()

data = c.execute('''SELECT * FROM junadata''')
retrieved_data = data.fetchall()
database_connection.commit()
database_connection.close()

retrieved_data[:10]
```
